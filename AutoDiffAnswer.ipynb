```
γ= Exp[I π kx] + 2*Exp[-I π kx/2]*Cos[ky π Sqrt[3]/2]
```

```
Dx = D[γ,kx];
Dy = D[γ,ky];
Dxx = D[γ,{kx,2}];
Dxy = D[γ,kx,ky];
Dyy = D[γ,{ky,2}];
```

```
plt = Table[ Dx , {ky,-1,1,1/64}, {kx,-1,1,1/64}];
real=ListDensityPlot[Re@plt] ;
imag=ListDensityPlot[Im@plt] ;
GraphicsRow[{real,imag}]
```
